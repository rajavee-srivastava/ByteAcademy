2. WAP to store student info in a JSON file - name, rollno, marks, grade, address - design your JSON carefully first and then display a menu to update the existing students details and add a new student detail to the file JSON file and do that as per the user's choice. Do it in MVC way.


```python
class Model:
    def __init__(self, filename):
        self.filename = filename
        self.data = {}
    
        if os.path.exists(self.filename):
            self.load_data()
        else:
            self.create_file()
    
    def load_data(self):
        with open(self.filename, 'r') as file:
            self.data = json.load(file)

    def create_file(self):
        with open(self.filename, 'w') as file:
            json.dump(self.data, file)
    
    def add_student(self, student:dict):
        self.data[student['roll_no']] = student
        with open(self.filename, 'w') as file:
            json.dump(self.data, file)

    def modify_student(self, roll_no, new_data:dict):
        for key, value in new_data.items:
            self.data[roll_no[key]] = value
        with open(self.filename, 'w') as file:
            json.dump(self.data, file)
    
            print()

    def show_all_students(self):
        for student in self.data.values():
            print("Name:", student['name'])
            print("Roll No:", student['roll_no'])
            print("Marks:", student['marks'])
            print("Grades:", student['grades'])
            print("Address:", student['address'])
            print()

class View:

    def new_student(self):
        print("Adding new student's info to the json file...")
        name = input("Enter student name: ")
        roll_no = input("Enter student's roll no: ")
        marks = input("Enter student marks: ")
        grade = input("Enter student grade: ")
        address = input("Enter student address:")
        student = {
            'name':name,
            'roll_no': roll_no,
            'marks': marks,
            'grades': grade,
            'address': address
        }
        return student
    
    def existing_student(self):
        updated_values = {'name': False, 'roll_no': False, 'marks': False, 'grades': False, 'address':False}
        roll_no = input("Enter Roll No of the student to update: ")
        new_data = {}
        
        fields = {
            'name': ('Name', 'Update Name?'),
            'email': ('Email', 'Update Email?'),
            'address': ('Address', 'Update Address?'),
            'date_of_birth': ('Date of Birth (YYYY-MM-DD)', 'Update Date of Birth?')
        }

        for field, (field_name, prompt) in fields.items():
            if input(prompt + " (y/n): ").lower() == 'y':
                value = input("Enter new {}: ".format(field_name))
                updated_values[field] = True
            else:
                value = None

            if updated_values[field]:
                new_data[field] = value
        
        return roll_no, new_data
    
class Controller:
    def __init__(self, model, view):
        self.model = model
        self.view = view

    def run(self):
        while True:
            print(
                """Welcome to the Student Management System!
            1. Add Student
            2. Update Student
            3. View Students
            4. Exit"""
            )

            choice = input("Enter your choice (1-4): ")

            if choice == '1':
                student = self.view.new_student()
                self.model.add_student(student)
                print("Student added successfully!")
                print()

            elif choice == '2':
                roll_no, new_data = self.view.existing_student()
                self.model.modify_student(roll_no, new_data)
                print("Student updated successfully!")
                print()

            elif choice == '3':
                print("All Students:")
                self.model.show_all_students()
            
            elif choice == '4':
                print("Exiting the program...")
                break
            
            else:
                print("Invalid choice. Please try again.")

# Example usage
filename = 'students.json'
model = Model(filename)
view = View()
controller = Controller(model, view)
controller.run()
```

5. Write an Algorithm(program) to Check if Two String are Anagram.

```python
word1 = input("Enter the first word: ")
word2 = input("Enter the second word: ")
dict1 = {letter: word1.count(letter) for letter in word1}
dict2 = {letter: word2.count(letter) for letter in word2}
if dict1 == dict2:
    print(word1, "and", word2, "are anagrams.")
else:
    print(word1, "and", word2, "are not anagrams.")
```
