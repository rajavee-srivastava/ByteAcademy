1. Write a function that takes a singly linked list and returns a complete copy of that list.

```python
class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self) -> None:
        self.head=None

    def insert(self, newNode):
        if self.head is None:
            self.head = newNode
        else:
            lastNode = self.head
            while True:
                if lastNode.next is None:
                    break
                lastNode = lastNode.next 
            lastNode.next = newNode

def printList(head):
    pointer = head
    while pointer is not None:
        print(pointer.data, end=' -> ')
        pointer = pointer.next
    print("None")

def copyList(head):
    if head is None:
        return None
    else:
        node = Node(head.data)
        node.next = copyList(head.next)
        return node
       

firstNode = Node("John")
linkedlist = LinkedList()
linkedlist.insert(firstNode)
secondNode = Node("Mike")
linkedlist.insert(secondNode)
thirdNode = Node("Amy")
linkedlist.insert(thirdNode)

copy = copyList(linkedlist.head)

printList(linkedlist.head)
printList(copy)
```

          the popped character is not the same as the current character, or it does not pair with the current 
          character of the expression. 
          Also, if the stack is empty, the total number of opening braces is less than the closing brace number
          at that point, so the expression cannot be balanced.   
                  

```python
brace_pairs = {'[':']', '(':')', '<':'>', '{':'}'}
stack = []

list_of_exp = ['{[{}{}]}[()]', '{{}{}}', '[]{}()', '{()}[)', '{(})', '][]{}}']
        
def balanced_expression(expressions: list): 
    """
    Check if the expression in the list are balanced or not. 
    """
    brace_pairs = {'[':']', '(':')', '<':'>', '{':'}'}
    stack = []
    
    for ex in expressions:
        print(ex)
        flag = False
        for char in ex:
            if char in brace_pairs.keys():
                stack.append(char)
            elif char in brace_pairs.values():
                pair_check = stack.pop()
                if brace_pairs[pair_check] == char:
                    continue
                else:
                    print("Not a balanced expression!")
                    flag =True
                    break
        if flag:
            pass
        elif len(stack) == 0:
            print("Balanced expression")

balanced_expression(list_of_exp)
```

3. Calculate the height of a binary tree. The program should consider the total number of nodes in the longest path. For example, an empty tree’s height is 0, and the tree’s height with only one node is 1.

```python
class TreeNode():
    def __init__(self, data):
        self.data = data
        self.children = []
        self.parent = None
    
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
        
    
    def get_level(self):
        level = 0
        p =self.parent
        while p:
            level +=1
            p = p.parent
        return level
    
    def printTree(self):
        spaces = " " * self.get_level() * 3
        prefix = spaces + "|--" if self.parent else ""
        print(prefix + self.data)
        if self.children:
            for child in self.children:
                child.printTree()
    
def height(root):
    """
    Calculate the height of a tree 
    """
    
    max_height = 0
    for child in root.children:
        max_height = max(max_height, height(child))
    return max_height + 1

root = TreeNode("0")

c1 = TreeNode("1")
c2 = TreeNode("2")
c3 = TreeNode("3")
c4 = TreeNode("4")

c2.add_child(c3)
c2.add_child(c4)

root.add_child(c1)
root.add_child(c2)

root.printTree()

height(root)
        
```

4. Graph coloring (also called vertex coloring) is a way of coloring a graph’s vertices such that no two adjacent 
   vertices share the same color. WAP to do it. Create a simple garph, assign colours and print it.


```python

# A class to represent a graph object
class Graph:
    def __init__(self, edges, n):
        self.adjList = [[] for _ in range(n)]
 
        # add edges to the undirected graph
        for (src, dest) in edges:
            self.adjList[src].append(dest)
            self.adjList[dest].append(src)
 
def colorGraph(graph, n):
 
    result = {}
 
    for u in range(n):
 
        assigned = set([result.get(i) for i in graph.adjList[u] if i in result])
 
        color = 1
        for c in assigned:
            if color != c:
                break
            color = color + 1
 
        result[u] = color
 
    for v in range(n):
        print(f'Color assigned to vertex {v} is {colors[result[v]]}')


colors = ['', 'BLUE', 'GREEN', 'RED', 'YELLOW', 'ORANGE', 'PINK',
        'BLACK', 'BROWN', 'WHITE', 'PURPLE', 'VOILET']

edges = [(0, 1), (0, 4), (0, 5), (4, 5), (1, 4), (1, 3), (2, 3), (2, 4)]

n = 6

graph = Graph(edges, n)
colorGraph(graph, n)
```

5. Implement a Queue using a linked List. 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def is_empty(self):
        return self.head is None
    
    def enqueue(self, data):
        node = Node(data)
        if self.is_empty():
            self.head = node
        else:
            self.tail.next = node
        self.tail = node
    
    def dequeue(self):
        if self.is_empty():
            raise Exception("Queue is empty")
        data = self.head.data
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        return data
    
queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print(queue.dequeue())  
queue.enqueue(4)
print(queue.dequeue())
```
